```python
#Re-formatting data so it only has average heartrate for each hour each day
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('heartrate.csv')
df = df.iloc[33:156]
df = df.iloc[:, 1:25]
def average_range(value):
    if isinstance(value, str) and '-' in value:
        try:
            low, high = map(float, value.replace(' ', '').split('-'))
            return (low+high)/2
        except:
            return np.nan
    else:
        return np.nan
df = df.applymap(average_range)
df.columns = range(df.shape[1])
df = df.reset_index(drop=True)
df.describe()
```

```python
#Plotting Hourly Average Heart Rate
hour_means = df.mean(axis=0)
plt.plot(hour_means)
plt.xlabel('Hour')
plt.ylabel('Average Heart Rate (bpm)')
plt.title('Average Heart Rate by Hour')
```

```python
#Selecting numerical values of calories burned from file
import pandas as pd
import numpy as np
df = pd.read_csv('activity.csv')
df = df.iloc[40:162]
df = df.iloc[:, 1:25]
df.columns = range(df.shape[1])
df = df.reset_index(drop=True)
#average calories burned per hour
calorie_means = df.mean(axis=0)
df.describe()
```

```python
#Plotting Average Calories Burned Per Hour
plt.plot(calorie_means)
plt.xlabel('Hour')
plt.ylabel('Calories Burned')
plt.title('Average Calories Burned')
```

```python
#Plotting Calories Burned vs Heart Rate
coefficients = np.polyfit(calorie_means, hour_means, 1)
poly = np.poly1d(coefficients)
hour_fit = poly(calorie_means)
correlation_matrix = np.corrcoef(calorie_means, hour_means)
r = correlation_matrix[0, 1]
r_squared = r**2
plt.legend(f'r^2={r**2}')
textstr = f'y = {coefficients[0]:.2f}x + {coefficients[1]:.2f}\nR² = {r_squared:.4f}'
plt.text(0.05, 0.95, textstr, transform=plt.gca().transAxes, 
         fontsize=10, verticalalignment='top',
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
plt.scatter(calorie_means, hour_means)
plt.plot(calorie_means, hour_fit, color='purple')
plt.title('Calories Burned vs Heart Rate')
plt.xlabel('Average Calories Burned')
plt.ylabel('Average Heart Rate (bpm)')
```
